#### Filter the data based on specific criteria and create a new filtered dataset

```python
subset_df = df[['race', 'background', 'justClass', 'alignment', 'level', 'HP', 'AC', 'Str', 'Dex', 'Con', 'Int', 'Wis', 'Cha']]
subset_df.head(20)
```

```python
subset_df = subset_df.dropna()
subset_df.size
```

```python
subset_df.head(20)
```

```python
#lets clean alignments
valid_alignments = ['LG', 'Lawful Good', 'NG', 'Neutral Good', 'CG', 'Chaotic Good',
                   'LN', 'Lawful Neutral', 'TN', 'True Neutral', 'CN', 'Chaotic Neutral',
                   'LE', 'Lawful Evil', 'NE', 'Neutral Evil', 'CE', 'Chaotic Evil']

subset_df = subset_df[subset_df['alignment'].isin(valid_alignments)]
```

```python
subset_df["alignment"].value_counts()
```

```python
valid_alignments = {
    'LG': 'Lawful Good', 'Lawful Good': 'Lawful Good',
    'NG': 'Neutral Good', 'Neutral Good': 'Neutral Good',
    'CG': 'Chaotic Good', 'Chaotic Good': 'Chaotic Good',
    'LN': 'Lawful Neutral', 'Lawful Neutral': 'Lawful Neutral',
    'TN': 'True Neutral', 'True Neutral': 'True Neutral',
    'CN': 'Chaotic Neutral', 'Chaotic Neutral': 'Chaotic Neutral',
    'LE': 'Lawful Evil', 'Lawful Evil': 'Lawful Evil',
    'NE': 'Neutral Evil', 'Neutral Evil': 'Neutral Evil',
    'CE': 'Chaotic Evil', 'Chaotic Evil': 'Chaotic Evil'
}

# Replace abbreviations with full names in the "alignment" column
subset_df['alignment'] = subset_df['alignment'].replace(valid_alignments)
subset_df["alignment"].value_counts()
```

```python
subset_df["justClass"].value_counts()
classes_dnd_5e = ['Artificer','Barbarian', 'Bard', 'Cleric', 'Druid', 'Fighter', 'Monk', 'Paladin', 'Ranger', 'Rogue', 'Sorcerer', 'Warlock', 'Wizard']
subset_df = subset_df[subset_df['justClass'].isin(classes_dnd_5e)]
subset_df["justClass"].value_counts()
```

```python
subrace_to_race = {
    'Hill Dwarf': 'Dwarf',
    'Mountain Dwarf': 'Dwarf',
    'High Elf': 'Elf',
    'Wood Elf': 'Elf',
    'Dark Elf (Drow)': 'Elf',
    'Lightfoot Halfling': 'Halfling',
    'Stout Halfling': 'Halfling',
    'Infernal Tiefling': 'Tiefling',
    'Abyssal Tiefling': 'Tiefling',
    'Protector Aasimar': 'Aasimar',
    'Scourge Aasimar': 'Aasimar',
    'Fallen Aasimar': 'Aasimar'
}

subset_df['race'] = subset_df['race'].apply(lambda x: subrace_to_race.get(x, x))
subset_df["race"].value_counts()
```

```python
original_races = ['Dwarf', 'Elf', 'Halfling', 'Human', 'Dragonborn', 'Gnome', 'Half-Elf', 'Half-Orc', 'Tiefling',
                  'Aarakocra', 'Genasi', 'Goliath', 'Bugbear', 'Firbolg', 'Kenku', 'Kobold', 'Lizardfolk',
                  'Tabaxi', 'Triton', 'Yuan-Ti Pureblood']

# Filter the DataFrame to include only the original race names
subset_df = subset_df[subset_df['race'].isin(original_races)]
subset_df["race"].value_counts()
```

#### Group the data by one or more columns and aggregate metrics like mean, max etc

```python
level_grouped = subset_df.groupby('level').mean().round(2)
level_grouped
```

```python
level_class_grouped = subset_df.groupby(['level', 'justClass']).mean().round(2)
specific_levels = [1, 5, 10, 20]
level_class_grouped = level_class_grouped[level_class_grouped.index.get_level_values('level').isin(specific_levels)]
level_class_grouped
```

#### Join the main dataset with another supplementary dataset

```python
supplementary = pd.read_csv('https://raw.githubusercontent.com/nogibjj/kh509_miniproject9/main/stats.csv', error_bad_lines=False)

supplementary.head(20)
```

```python
proper_race_names = {
    'dragonborn': 'Dragonborn',
    'halfling': 'Halfling',
    'gnome': 'Gnome',
    'human': 'Human',
    'tiefling': 'Tiefling',
    'elf': 'Elf',
    'dwarf': 'Dwarf',
    'half.elf': 'Half-Elf',
    'half.orc': 'Half-Orc'
}

# Rename specified column observations with proper race names
supplementary.replace(proper_race_names, inplace=True)
supplementary.head(5)
```

```python
# Drop unnecessary columns
supplementary.drop(['height', 'weight', 'speed'], axis=1, inplace=True)
# Rename columns
supplementary.rename(columns={'strength': 'Str', 'dexterity': 'Dex', 'constitution': 'Con',
                              'intelligence': 'Int', 'wisdom': 'Wis', 'charisma': 'Cha'}, inplace=True)
supplementary['background'] = ''
supplementary['justClass'] = ''
supplementary['alignment'] = ''
supplementary['level'] = ''
supplementary['HP'] = ''
supplementary['AC'] = ''
merged_df = pd.concat([subset_df, supplementary], ignore_index=True)

merged_df.reset_index(drop=True, inplace=True)
```

#### Create various data visualizations like histograms, scatter plots, box plots etc

```python
import seaborn as sns
import matplotlib.pyplot as plt

level_1_df = merged_df[merged_df['level'] == 1]

# List of statistics you want to create boxplots for
statistics = ['Str', 'Dex', 'Con', 'Int', 'Wis', 'Cha']

# Loop through each statistic and create boxplots for level 1 characters
for stat in statistics:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='justClass', y=stat, data=level_1_df)
    plt.title(f'Boxplot of {stat} by Class (Level 1)')
    plt.xlabel('Class')
    plt.ylabel(stat)
    plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
    plt.show()
```
